```python
from ztfquery import lightcurve
import matplotlib.pyplot as plt

n = 'Mrk 817'
ra = 219.0920037 
dec = 58.7942865

datar = lightcurve.LCQuery.download_data(circle=[ra,dec,0.001], bandname="r")#, MAG= [17.4, 18])
lcrr = lightcurve.LCQuery(datar)

datag = lightcurve.LCQuery.download_data(circle=[ra,dec,0.001], bandname="g")#,, MAG= [17.4, 18])
lcgg = lightcurve.LCQuery(datag)

datai = lightcurve.LCQuery.download_data(circle=[ra,dec,0.001], bandname="i")#,, MAG= [17.4, 18]) 
lcii = lightcurve.LCQuery(datai)
```

```python
timer = datar['mjd']
magr = datar['mag']
magerrr = datar['magerr']

timeg = datag['mjd']
magg = datag['mag']
magerrg = datag['magerr']

timei = datai['mjd']
magi = datai['mag']
magerri = datai['magerr']

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=(10, 13))

# Plot the magnitude graph with error bars
ax1.errorbar(timer, magr, yerr=magerrr, fmt='o',ecolor='darkgrey',markeredgecolor='darkgrey',markersize =4, color='r')
ax2.errorbar(timeg, magg, yerr=magerrg, fmt='o',ecolor='darkgrey',markeredgecolor='darkgrey',markersize =4, color='g')
ax3.errorbar(timei, magi, yerr=magerri, fmt='o',ecolor='darkgrey',markeredgecolor='darkgrey',markersize =4, color='darkorange')

ax1.set_xlabel('MJD')
ax1.set_ylabel('Magnitude')
ax1.legend("r-band")
ax1.invert_yaxis()
ax1.set_title((n,ra,dec))

ax2.set_xlabel('MJD')
ax2.set_ylabel('Magnitude')
ax2.legend('g-band')
ax2.invert_yaxis()

ax3.set_xlabel('MJD')
ax3.set_ylabel('Magnitude')
ax3.legend('i-band')
ax3.invert_yaxis()

#plt.savefig(n) #save fig 
plt.show()
```
